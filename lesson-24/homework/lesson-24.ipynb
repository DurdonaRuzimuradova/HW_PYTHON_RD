```python
# **Homework Assignment 1: Analyzing Sales Data**

# You are given a dataset containing sales data for an e-commerce website. 
# The dataset (`task\sales_data.csv`) has the following columns:

# - `Date`: Date of the sale.
# - `Product`: Name of the product sold.
# - `Category`: Category to which the product belongs.
# - `Quantity`: Number of units sold.
# - `Price`: Price per unit.

import pandas as pd
df=pd.read_csv(r'C:\Users\hp\Desktop\Maab_Academy\Python\HW_24\task\sales_data.csv')
df.head()

```

```python
# **Tasks:**
# 1. Group the data by the `Category` column and calculate the following aggregate statistics for each category:
#    - Total quantity sold.
#    - Average price per unit.
#    - Maximum quantity sold in a single transaction.
(df
 .groupby('Category')
 .agg({'Quantity': ['sum','max'],
       'Price':'mean'}).round(2)
)
```

```python
# 2. Identify the top-selling product in each category based on the total quantity sold.
dff=(df
     .groupby(['Category','Product'])
     .agg({'Quantity':'sum'})
     .reset_index())
top_salling=dff.loc[dff.groupby('Category')['Quantity'].idxmax()]
top_salling
```

```python
# 3. Find the date on which the highest total sales (quantity * price) occurred.
df['Total']=df.Quantity*df.Price
df1=(df
     .groupby('Date')
     .agg({'Total':'sum'})
     .reset_index()
    )
max_sales_date=df1.loc[df1.Total.idxmax()]
max_sales_date
```

```python
# **Homework Assignment 2: Examining Customer Orders**

# You have a dataset (`task\customer_orders.csv`) containing information about customer orders. The dataset has the following columns:

# - `OrderID`: Unique identifier for each order.
# - `CustomerID`: Unique identifier for each customer.
# - `Product`: Name of the product ordered.
# - `Quantity`: Number of units ordered.
# - `Price`: Price per unit.
df=pd.read_csv(r'C:\Users\hp\Desktop\Maab_Academy\Python\HW_24\task\customer_orders.csv')
df.head()
```

```python

# **Tasks:**

# 1. Group the data by `CustomerID` and filter out customers who have made less than 20 orders.
df1=(df
    .groupby('CustomerID')
    .agg({'OrderID':'count'})
    .reset_index()
    )
filtered_df=df1[df1['OrderID']>=20]
filtered_df
```

```python
# 2. Identify customers who have ordered products with an average price per unit greater than $120.
df2=(df
    .groupby('CustomerID')
    .agg({'Price':'mean'})
    .reset_index()
    )
filtered_df=df2[df2.Price>120]
filtered_df
```

```python
# 3. Find the total quantity and total price for each product ordered, and filter out products that have a total quantity less than 5 units.
df2=(df
    .groupby('Product')
    .agg({'Quantity':'sum',
        'Price':'sum'})
    .reset_index()
    )
filtered_df=df2[df2.Quantity>=5]
filtered_df
```

```python
# **Homework Assignment 3: Population Salary Analysis**

# 1. "task\population.db" sqlite database has `population` table.

# 2. "task\population salary analysis.xlsx" file defines Salary Band categories. <br />

import sqlite3
import pandas as pd
db_file_path = r'C:\Users\hp\Desktop\Maab_Academy\Python\HW_24\task\population.db'
connection = sqlite3.connect(db_file_path)
cursor = connection.cursor()
cursor.execute("SELECT * FROM population")
rows = cursor.fetchall()
column_names = [description[0] for description in cursor.description]
df1= pd.DataFrame(rows, columns=column_names)
cursor.close()
connection.close()
print(df1.head())
print(df1.columns)

# Note: Use SQL only to select data from database. All the other calculations should be done in python.
```

```python
#     Read the data from population table and calculate following measures:

import pandas as pd

# Specify the path to your Excel file
file_path = r'C:\Users\hp\Desktop\Maab_Academy\Python\HW_24\task\population_salary_analysis.xlsx'

# Read the Excel file into a DataFrame
dff = pd.read_excel(file_path)

# Display the DataFrame
print(dff['Salary Band'])
```

```python
#     - Percentage of population for each salary category;
def assign_salary_band(salary):
    if salary <= 200000:
        return 'till $200,000'
    elif salary <= 400000:
        return '$200,001 - $400,000'
    elif salary <= 600000:
        return '$400,001 - $600,000'
    elif salary <= 800000:
        return '$600,001 - $800,000'
    elif salary <= 1000000:
        return '$800,001 - $1,000,000'
    elif salary <= 1200000:
        return '$1,000,001 - $1,200,000'
    elif salary <= 1400000:
        return '$1,200,001 - $1,400,000'
    elif salary <= 1600000:
        return '$1,400,001 - $1,600,000'
    elif salary <= 1800000:
        return '$1,600,001 - $1,800,000'
    else:
        return '$1,800,001 and over'

df1['Salary Band'] = df1['salary'].apply(assign_salary_band)
population_count = df1['Salary Band'].value_counts()
total_population = len(df1)
percentage_population = (population_count / total_population) * 100
percentage_df = pd.DataFrame(percentage_population).reset_index()
percentage_df.columns = ['Salary Band', 'Population Percentage']
print(percentage_df)
```

```python
#     - Average salary in each salary category;
#     - Median salary in each salary category;
#     - Number of population in each salary category;
df2 = df1.groupby('Salary Band').agg(
    Average_Salary=('salary', 'mean'),
    Median_Salary=('salary', 'median'),
    Population_Count=('salary', 'count')
).reset_index()
print(df2)
# 3. Calculate the same measures in each State
```
